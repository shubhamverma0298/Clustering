#**Theoretical Questions**

**Q 1 What is unsupervised learning in the context of machine learning?**
  - In machine learning the meaning of the unsupervised learning is that it dosen't use the Y(target variable ) to make the machine learning model.

**Q 2 How does K-Means clustering algorithm work?**
  - The algorithm iteratively assigns data points to the nearest cluster centroid and then recalculates the centroids based on the assigned points, repeating this process until convergence.

**Q 3 Explain the concept of a dendrogram in hierarchical clustering?**
  - A dendrogram is a tree-like diagram that visualizes the hierarchical relationships between data points or clusters, generated by hierarchical clustering.

**Q 4 What is the main difference between K-Means and Hierarchical Clustering?**
  - The primary difference between K-means and hierarchical clustering lies in how they form clusters. K-means requires a pre-defined number of clusters (k) and iteratively assigns data points to the nearest centroid, while hierarchical clustering builds a tree-like structure (dendrogram) showing how data points are merged or split, without needing a predetermined number of clusters.

**Q 5 What are the advantages of DBSCAN over K-Means?**
  - DBSCAN can identify clusters of arbitrary shape, is more robust to noise and outliers, and doesn't require specifying the number of clusters beforehand.
  - DBSCAN offers several advantages over K-Means, particularly when dealing with datasets containing irregularly shaped clusters, outliers, or varying densities.

**Q 6 When would you use Silhouette Score in clustering?**
  - The silhouette score is used to evaluate the quality of clustering results, specifically to assess how well data points are assigned to their respective clusters.

**Q 7 What are the limitations of Hierarchical Clustering?**
  - Hierarchical clustering, despite its strengths, faces limitations like high computational cost for large datasets, sensitivity to noise and outliers, and the inability to revisit merging decisions.

**Q 8 Why is feature scaling important in clustering algorithms like K-Means?**
  - Feature scaling is crucial for k-means clustering because it ensures that all features contribute equally to the distance calculations, preventing features with larger scales from dominating the process.

**Q 9 How does DBSCAN identify noise points?**
  - The algorithm works by defining clusters as dense regions separated by regions of lower density.

**Q 10 Define inertia in the context of K-Means?**
  - Inertia is calculated by measuring the distance between each data point and its centroid, squaring this distance, and summing these squares across one cluster.

**Q 11 What is the elbow method in K-Means clustering?**
  - The elbow method is a heuristic used in K-means clustering to find the optimal number of clusters (k) for a dataset.

**Q 12 Describe the concept of "density" in DBSCAN?**
  - In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), density refers to the concentration of data points within a specified radius (epsilon, ε) around a given point.

**Q 13 Can hierarchical clustering be used on categorical data?**
  - Yes, hierarchical clustering can be used with categorical data.

**Q 14 What does a negative Silhouette Score indicate?**
  - Negative values generally indicate that a sample has been assigned to the wrong cluster.

**Q 15 Explain the term "linkage criteria" in hierarchical clustering?**
  - In hierarchical clustering, linkage criteria define how the distance between clusters is calculated when merging them.

**Q 16 Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?**
  - K-Means clustering can struggle with datasets exhibiting varying cluster sizes or densities because it assumes all clusters are roughly spherical, equally sized, and equally dense.

**Q 17 What are the core parameters in DBSCAN, and how do they influence clustering?**
  - DBSCAN, a density-based clustering algorithm, relies on two core parameters: **epsilon (ε or eps) and MinPts**.

**Q 18 How does K-Means++ improve upon standard K-Means initialization?**
  - K-Means++ improves upon standard K-Means by using a more intelligent initialization method for cluster centroids.

**Q 19 What is agglomerative clustering?**
  - Agglomerative clustering is a hierarchical clustering technique that builds a hierarchy of clusters in a "bottom-up" manner.

**Q 20 What makes Silhouette Score a better metric than just inertia for model evaluation?**
  - The Silhouette Score is generally a better metric than inertia for evaluating clustering models because it considers both cohesion (how tightly packed a cluster is) and separation (how well-separated clusters are).

**Q 1 Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot?**

```python
from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=1000,n_features=2,centers=4,random_state=1)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
from sklearn.cluster import KMeans
model=KMeans(n_clusters=4,random_state=1)
model.fit(X_train)
y_labels=model.predict(X_train)
import matplotlib.pyplot as plt
plt.scatter(X_train[:,0],X_train[:,1],c=y_labels)
plt.title("K-Means Clustering with 4 centers in data ")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

**Q 2 Load the Iris dataset and use Agglomerative Clustering to group the data into 3 clusters. Display the first 10
predicted labels?**

```python
from sklearn.datasets import load_iris
data=load_iris()
X=data.data
y=data.target
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X)
X_train
from sklearn.decomposition import PCA
pca=PCA(n_components=3)
pca_scalled=pca.fit_transform(X_train)
from sklearn.cluster import AgglomerativeClustering
ago=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
labels=[]
for i in range(1,10):
    ago=AgglomerativeClustering(n_clusters=i,metric='euclidean',linkage='ward')
    ago.fit_predict(pca_scalled)
    label=ago.labels_
    labels.append(label)
labels
```

**Q 9 Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart?**
